## Import the relevant libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
sns.set()
```

## Load the data

Load data from the csv file: <i> 'Countries.csv'</i>.


```python
df = pd.read_csv('Countries-exercise.csv')
df
```

## Plot the data

Plot the <i>'Longtitude'</i> and <i>'Latitude'</i> columns. 

```python
plt.scatter(df['Longitude'], df['Latitude'])
plt.xlim(-180, 180)
plt.ylim(-90, 90)
plt.show()
```

## Select the features

Create a copy of that data and remove all parameters apart from <i>Longitude</i> and <i>Latitude</i>.

```python
x = df.iloc[:, 1:3]
x
```

## Clustering

Assume there are only two clusters. 

```python
kmeans = KMeans(2) #Cluster with k value=2
```

```python
kmeans.fit(x)
```

### Clustering Resutls

```python
identified_results = kmeans.fit_predict(x)
identified_results
```

```python
data_with_clusters = df.copy()
data_with_clusters['Cluster'] = identified_results
data_with_clusters
```

Did you remember to use the <i> c </i> parameter to separate the data by the clusters we defined?

```python
plt.scatter(data_with_clusters['Longitude'], data_with_clusters['Latitude'], c=data_with_clusters['Cluster'], cmap='rainbow')
plt.xlim(-180, 180)
plt.ylim(-90, 90)
plt.show()
```

***
***
There is a method to choose the perfect clusters is know as **Elbow Method**

The distance between points in a cluster is known as **With-in cluster sum of squares** or **WCSS**

If we minimize the _WCSS_ the we have reached the perfect clustering solution, but remember _WCSS_ cannot be 0

```python
#WCSS:
kmeans.inertia_
```

```python
# lets for wcss for 1 to 8:
wcss = []
for i in range(1,9):
    kmeans = KMeans(i)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
wcss
```

The above sequence is decreasing with very big leaps in the first two steps:

Lets plot a graph and check for the elbow method:

```python
number_clusters = range(1,9)
plt.plot(number_clusters, wcss)
plt.title('The Elbow Method')
plt.xlabel('no. of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
wcss = []
for i in range(1,101, 10):
    kmeans = KMeans(i)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)


number_clusters = range(1,101, 10)
plt.plot(number_clusters, wcss)
plt.title('The Elbow Method')
plt.xlabel('no. of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# From the above plot it shows the perfect cluster number is 10
kmeans = KMeans(10) #10 clusters
kmeans.fit(x)

identified_clusters = kmeans.fit_predict(x)
data_with_clusters = df.copy()
data_with_clusters['Cluster'] = identified_clusters

plt.scatter(data_with_clusters['Longitude'], data_with_clusters['Latitude'], c=data_with_clusters['Cluster'], cmap='rainbow')
plt.xlim(-180, 180)
plt.ylim(-90, 90)
plt.show()
```
